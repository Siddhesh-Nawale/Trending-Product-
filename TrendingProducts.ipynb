```python
import pandas as pd
import numpy as np
import operator

SMOOTHING_WINDOW_FUNCTION = np.hamming
SMOOTHING_WINDOW_SIZE = 7
#df = pd.read_excel('sample-cart-add-data.xlsx')
def TrendingProductID(df):
    #df = pd.read_excel('sample-cart-add-data.xlsx')
    df.sort_values(by=['ProductId', 'age'], inplace=True)
    trends = pd.pivot_table(df, values='count', index=['ProductId', 'age'])

    trend_snap = {}

    for i in np.unique(df['ProductId']):
        trend = np.array(trends.loc[i])
        smoothed = smooth(trend, SMOOTHING_WINDOW_SIZE, SMOOTHING_WINDOW_FUNCTION)
        nsmoothed = standardize(smoothed)
        slopes = nsmoothed[1:] - nsmoothed[:-1]
        # I blend in the previous slope as well, to stabalize things a bit and
        # give a boost to things that have been trending for more than 1 day
        if len(slopes) > 1:
            trend_snap[i] = slopes[-1] + slopes[-2] * 0.5
    return sorted(trend_snap.items(), key=operator.itemgetter(1), reverse=True)


def CategoryID(df):
    #df = pd.read_excel('sample-cart-add-data.xlsx')
    df.sort_values(by=['CategoryID', 'age'], inplace=True)
    trends = pd.pivot_table(df, values='count', index=['CategoryID', 'age'])

    trend_snap = {}

    for i in np.unique(df['CategoryID']):
        trend = np.array(trends.loc[i])
        smoothed = smooth(trend, SMOOTHING_WINDOW_SIZE, SMOOTHING_WINDOW_FUNCTION)
        nsmoothed = standardize(smoothed)
        slopes = nsmoothed[1:] - nsmoothed[:-1]
        # I blend in the previous slope as well, to stabalize things a bit and
        # give a boost to things that have been trending for more than 1 day
        if len(slopes) > 1:
            trend_snap[i] = slopes[-1] + slopes[-2] * 0.5
    return sorted(trend_snap.items(), key=operator.itemgetter(1), reverse=True)


def MaterialID(df):
    #df = pd.read_excel('sample-cart-add-data.xlsx')
    df.sort_values(by=['MaterialID', 'age'], inplace=True)
    trends = pd.pivot_table(df, values='count', index=['MaterialID', 'age'])

    trend_snap = {}

    for i in np.unique(df['MaterialID']):
        trend = np.array(trends.loc[i])
        smoothed = smooth(trend, SMOOTHING_WINDOW_SIZE, SMOOTHING_WINDOW_FUNCTION)
        nsmoothed = standardize(smoothed)
        slopes = nsmoothed[1:] - nsmoothed[:-1]
        # I blend in the previous slope as well, to stabalize things a bit and
        # give a boost to things that have been trending for more than 1 day
        if len(slopes) > 1:
            trend_snap[i] = slopes[-1] + slopes[-2] * 0.5
    return sorted(trend_snap.items(), key=operator.itemgetter(1), reverse=True)



def smooth(series, window_size, window):
    ext = np.r_[2 * series[0] - series[window_size-1::-1],
                series,
                2 * series[-1] - series[-1:-window_size:-1]]
    ext=ext[:,0]
    weights = window(window_size)
    smoothed = np.convolve(weights / weights.sum(), ext, mode='same')
    return smoothed[window_size:-window_size+1]

 
def standardize(series):
    import numpy as np
    #series = series[~np.isnan(series)]
    iqr = np.quantile(series,0.75) - np.quantile(series,0.25)
    return (series - np.median(series)) / iqr

```
